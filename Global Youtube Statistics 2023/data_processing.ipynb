```python
# import pandas 
import pandas as pd
```

```python
# read data from the file and handle null values 
df = pd.read_csv('./data/Global YouTube Statistics.csv', na_values=['Not Available','unknown'], encoding='latin-1')
```

```python
# checking NaN values in the dataframe 
df.isna().sum().sum()
```

```python
# checking NaN values in every column
df.columns[df.isna().any()]
```

```python
# checking NaN values in data frame 
nan_counts = df.isna().sum()
nan_counts
```

```python
# drop rows with any NaN values in the dataframe
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# re-checking NaN values 
nan_counts = df.isna().sum()
```

```python
nan_counts
```

```python
# check data type
print(df.dtypes)
```

```python
# Based on the data types, the columns that need to be formatted correctly are:
# created_year: This should be an integer type.
df['created_year'] = df['created_year'].astype(int)
# created_date: This should be a datetime type.
df['created_date'] = pd.to_datetime(df['created_date'])
```

```python
# check data type
print(df.dtypes)
```

```python
# save to cleaned dataframe to a new CSV file
df.to_csv('./data/cleaned_youtube_data.csv', index=False)
print(("CSV file 'cleaned_youtube_data.csv' created succesfully!"))
```
